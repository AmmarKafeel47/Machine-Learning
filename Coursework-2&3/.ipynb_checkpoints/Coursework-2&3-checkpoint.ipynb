## 1. Import Necessary Libraries
Importing all the necessary libraries that will help me in building the model.


```python
#Importing all the necessary libraries that will help me in building the model.
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import plot_model
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
```

## 2. Public Dataset 
[1] Cortinhas, S. (2020). Apples or Tomatoes Image Classification. Kaggle. https://www.kaggle.com/datasets/samuelcortinhas/apples-or-tomatoes-image-classification

Importing the dataset for my model.


```python
# Setting the paths to the training and testing folders
train_dir = r'C:\Users\ammar\Desktop\New folder\Machine Learning\Coursework-2&3\DataSet\train'
test_dir = r'C:\Users\ammar\Desktop\New folder\Machine Learning\Coursework-2&3\DataSet\test'
```

## 3. Defining Parameters

```python
# Defining the batch size and the image size..
batch_size = 32
img_size = (224, 224)
```

## 4. Data Generators


```python
# Defining the data generators for training and testing sets
train_data_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
test_data_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
 
train_generator = train_data_gen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary')

test_generator = test_data_gen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary')
```

## 5. Splitting Data

```python
# Spliting the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    train_generator.filenames,
    train_generator.classes,
    test_size=0.2,
    shuffle=False, # do not shuffle to preserve the order of the samples
    random_state=42,
    stratify=None # the class distribution is already balanced
)


# Converting the output of train_test_split to numpy arrays
#The output of train_test_split is in the form of lists, 
#so I am converting it to numpy arrays using np.array.

X_train = np.array(X_train)
X_val = np.array(X_val)
y_train = np.array(y_train)
y_val = np.array(y_val)

# Now printing the shape of the training and validation sets
print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
```

## 6. Defining Input Shape and Number of Classes

```python
# Defining the input shape
input_shape = (224, 224, 3)

# Defining the number of classes
num_classes = 2
```

## 7. Designing a Neural Network
I am using a convolutional neural network, this code trains a model to categorise images of apples and tomatoes. Convolutional and dense layers with dropout are used to build the model, which is then built using binary cross-entropy loss and the Adam optimizer. To avoid overfitting, early stopping is used as an additional feature, and accuracy is used to evaluate the model. With a test accuracy of roughly 90%, the final model can discriminate between the 2 classes. More brief explaination is provided in the pdf, respectively.

```python
# Defining the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(256, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compiling the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Now to visualize the model structure
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
model.summary()
```

## 8. Additional Feature 1: Visualizing Test Images

```python
import matplotlib.pyplot as plt
# Get a batch of images from the test generator
images, labels = next(test_generator)
# Display a few images
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i])
    plt.title(labels[i])
    plt.axis('off')
plt.show()
```

## 9. Additional Feature 2: Early Stopping

Im adding another extra feature to the model: early stopping. Early stopping is a strategy for preventing overfitting by assessing the models performance on a validation set during training. If the validation set performance stops increasing, the training process is terminated early to avoid the model from remembering the training data and becoming less successful at generalizingto new data.

```python
# Defining the early stopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

# Training the model with early stopping
history = model.fit(
    train_generator,
    epochs=20,
    validation_data=test_generator,
    callbacks=[early_stopping]
)
# Evaluating the model using the testing dataset
test_loss, test_acc = model.evaluate(test_generator, verbose=2)
print('Test accuracy:', test_acc)
```

```python
#********************************************  PRESENTING THE CLASSIFICATION RESULTS  *******************************************
img = load_img(root.filename, target_size=img_size)
img_array = img_to_array(img)
img_array = img_array.reshape((1,) + img_array.shape)
img_array = img_array / 255.0

prediction = model.predict(img_array)
if prediction[0][0] > 0.5:
    print('Prediction: TOMATO')
else:
    print('Prediction: APPLE')
```
